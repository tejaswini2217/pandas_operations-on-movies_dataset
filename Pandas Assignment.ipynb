```python

import pandas as pd
df=pd.read_excel("Movies_Dataset.xlsx")
df
```

```python
# 1. Select all movies in the "Action" genre.
movies=df[df["Genre"] == "Action"]
movies
```

```python
#2. Select movies released after 2000. 
movies=df[df["Release_Year"] > 2000]
print(movies)
```

```python
#3. Select movies with an IMDb rating greater than 8.5. 
movies=df[df["IMDb_Rating"] > 8.5]
movies
```

```python
#4. Select only "Movie Name" and "Genre" columns. 
df[["Movie Name","Genre"]]
```

```python
#5. Select movies released between 1990 and 2010. 
movies=df[(df["Release_Year"] > 1990) & (df["Release_Year"] < 2010)]
movies
```

```python
#6. Count the number of movies in each genre. 
df["Genre"].value_counts()
```

```python
#7. Find the highest IMDb-rated movie. 
df[df["IMDb_Rating"]==df.IMDb_Rating.max()]
```
